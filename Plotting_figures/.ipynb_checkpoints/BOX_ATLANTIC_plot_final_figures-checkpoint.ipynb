## for the bar chart

```python
 # construct list for the bar chart
def list_var(nexp,var_name):
    dir_base = dir_work + "/ETUDE3/ENS_VAR_BUDGET/data4analyses/"
    var_SLA = xr.open_zarr(dir_base+nexp+"/SLA/var_terms_box_atl.zarr")
    var_ste = xr.open_zarr(dir_base+nexp+"/"+"STERIC"+"/var_terms_box_atl_0_bottom.zarr")
    var_the = xr.open_zarr(dir_base+nexp+"/"+"THERMO"+"/var_terms_box_atl_0_bottom.zarr")
    var_hal = xr.open_zarr(dir_base+nexp+"/"+"HALO"+"/var_terms_box_atl_0_bottom.zarr")
    var_man = xr.open_zarr(dir_base+nexp+"/"+"MANO"+"/var_terms_box_atl.zarr")
    return [var_SLA.variances.sel(var_term=var_name).values,\
           var_ste.variances.sel(var_term=var_name).values,\
           var_the.variances.sel(var_term=var_name).values,\
           var_hal.variances.sel(var_term=var_name).values,\
           var_man.variances.sel(var_term=var_name).values]
```

```python
list_tot_var1 = list_var("ES",'tot_var')
list_forced_var1 = list_var("ES",'forced_var')
list_intrinsic_var1 = list_var("ES",'intrinsic_var')
list_res_var1 = list_var("ES",'res_var')

list_tot_var2 = list_var("EAI",'tot_var')
list_forced_var2 = list_var("EAI",'forced_var')
list_intrinsic_var2 = list_var("EAI",'intrinsic_var')
list_res_var2 = list_var("EAI",'res_var')
```

```python
def plot_bar_chart(ax):
    # set width of bar 
    barWidth = 0.10
    
    # Set position of bar on X axis 
    # PLOT BAR OF --------------------------- ES
    # Set position of bar on X axis 
    br1 = np.arange(len(list_tot_var1)) 
    br2 = [x + 2*barWidth for x in br1] 
    br3 = [x + 2*barWidth for x in br2] 
    br4 = [x + 2*barWidth for x in br3] 
                                                 
    # Make the plot
    b1 = ax.bar(br1, list_tot_var1, color ="k", width = barWidth, edgecolor ='grey', label ='Total') 
    b2 = ax.bar(br2, list_forced_var1, color =CB_colors['blue'], width = barWidth, edgecolor ='grey', label ='Forced') 
    b3 = ax.bar(br3, list_intrinsic_var1, color ='darkgoldenrod', width = barWidth, edgecolor ='grey', label ='Intrinsic') 
    b4 = ax.bar(br4, list_res_var1, color ='gold', width = barWidth, edgecolor ='grey', label ='Bias') 


    # PLOT BAR OF --------------------------- EAI
    # Set position of bar on X axis 
    br1 = np.arange(len(list_tot_var2)) + barWidth
    br2 = [x + 2*barWidth for x in br1] 
    br3 = [x + 2*barWidth for x in br2] 
    br4 = [x + 2*barWidth for x in br3] 
    
    # Make the plot
    b1_bis = ax.bar(br1, list_tot_var2,     color ="k", width = barWidth, alpha=0.7,edgecolor ='grey') 
    b2_bis = ax.bar(br2, list_forced_var2, color =CB_colors['blue'], width = barWidth,alpha=0.5,edgecolor ='grey') 
    b3_bis = ax.bar(br3, list_intrinsic_var2, color ='darkgoldenrod', width = barWidth, alpha=0.5,edgecolor ='grey') 
    b4_bis = ax.bar(br4, list_res_var2,      color ='gold', width = barWidth, alpha=0.5,edgecolor ='grey') 


    ax.set_yticks(np.arange(0,1.6,0.2)) 

    ax.set_xticks([r + barWidth for r in range(len(list_tot_var1))]) 
    ax.set_xticklabels(['a', 'b', 'd', 'e', 'c']) #, fontsize=12) ,rotation=45, rotation_mode="anchor"

    # plt.ylabel(r"Variance [$cm^{2}$]")
    # plt.xticks([r + barWidth for r in range(len(list_tot_var1))], ['SLA', 'Steric', 'Thermosteric', 'Halosteric']) 

    l = ax.legend([(b1, b1_bis),(b2, b2_bis),(b3, b3_bis),(b4, b4_bis)], ['Total ES|ER', 'Forced ES|ER', 'Intrinsic ES|ER', 'Bias ES|ER'],\
               handler_map={tuple: HandlerTuple(ndivide=None)})
```

## for the time series

```python
def load_plot_ts_ssh_mano(nexp,COMP,ax):
    dir_base = dir_work + "/ETUDE3/ENS_VAR_BUDGET/data4analyses/"+nexp+"/"

    ts_box_atl = xr.open_zarr(dir_base+COMP+"/time_series_box_atl.zarr").ts_box_atl
    
    for ii in np.arange(0,10):
        (ts_box_atl.isel(ens=ii)).plot( color='grey', ax=ax)
        
    ts_box_atl.mean('ens').plot(color=CB_colors['blue'],ax=ax)
    ax.grid()
    lim1 = -3
    lim2 = 3
    ax.set_ylim(lim1,lim2)
    ax.set_ylabel(None)
    ax.set_xlabel("Time")

    return ts_box_atl



def load_plot_tsCOMP(nexp,COMP,ax):
    dir_base = dir_work + "/ETUDE3/ENS_VAR_BUDGET/data4analyses/"+nexp+"/"

    test = xr.open_zarr(dir_base+"SLA"+"/time_series_box_atl.zarr").ts_box_atl
    ts_box_atl = xr.open_zarr(dir_base+COMP+"/time_series_box_atl_0_bottom.zarr").ts_box_atl.assign_coords({"year":test.time_counter.values})
    
    for ii in np.arange(0,10):
        (ts_box_atl.isel(ens=ii)).plot( color='grey', ax=ax)
        
    ts_box_atl.mean('ens').plot(color=CB_colors['blue'],ax=ax)
    ax.grid()
    lim1 = -3
    lim2 = 3
    ax.set_ylim(lim1,lim2)
    ax.set_ylabel(None)
    ax.set_xlabel("Time")
    return ts_box_atl
```

## PLOT FIGURE 1

```python
cm = 1/2.54  # centimeters in inches

nexp="ES"

fig, [(ax0,ax1,ax2), (ax3,ax4,ax5)] = plt.subplots(2,3,figsize=(35*cm, 15*cm), facecolor='none', dpi=72*2) # 
plt.subplots_adjust(wspace=0.1, hspace=0.5)

load_plot_ts_ssh_mano(nexp,"SLA",ax0)
load_plot_tsCOMP(nexp,"STERIC",ax1)
load_plot_ts_ssh_mano(nexp,"MANO",ax2)
load_plot_tsCOMP(nexp,"THERMO",ax3)
load_plot_tsCOMP(nexp,"HALO",ax4)

ax0.set_title("a) SLA")
ax1.set_title("b) Steric")
ax2.set_title("c) Manometric")
ax3.set_title("d) Thermosteric")
ax4.set_title("e) Halosteric")
ax5.set_title(r"f) Variances [$cm^{2}$]")

# ax0.legend(['Total',"Force","Internal", "Verification"], loc="lower right", fontsize=8)
ax0.set_ylabel('[cm]')
ax3.set_ylabel('[cm]')
# plt.xlabel('Time'); 

# bar chart
plot_bar_chart(ax5)

namo="fig_synthetic_atlantic_box_time_series.png"
# plt.show()
fig.savefig(diro+namo, facecolor=fig.get_facecolor(), bbox_inches='tight', transparent=True) #,edgecolor='none', bbox_inches='tight'
# plt.close(fig)
```

```python
dir_base = dir_work + "/ETUDE3/ENS_VAR_BUDGET/data4analyses/"

def list_var_dep(nexp,COMP,var_name):
    return_list=[]
    
    for dep in ["50m", "100m", "300m", "700m", "2000m", "bottom"]: 
        var_ES_dep = xr.open_zarr(dir_base+nexp+"/"+COMP+"/var_terms_box_atl_0_"+dep+".zarr").variances
        return_list.append(var_ES_dep.sel(var_term=var_name).values)

    return return_list
```

```python
# example of a bar chart
def plot_bar_chart_dep(COMP,ax):
    # set width of bar 
    barWidth = 0.10
    
    # PLOT BAR OF --------------------------- ES
    # set height of bar 
    list_tot_var1 = list_var_dep('ES',COMP,'tot_var')
    list_forced_var1 = list_var_dep('ES',COMP,'forced_var')
    list_intrinsic_var1 = list_var_dep('ES',COMP,'intrinsic_var')
    list_res_var1 = list_var_dep('ES',COMP,'res_var')

    # Set position of bar on X axis 
    br1 = np.arange(len(list_tot_var1)) 
    br2 = [x + 2*barWidth for x in br1] 
    br3 = [x + 2*barWidth for x in br2] 
    br4 = [x + 2*barWidth for x in br3] 
                                                 
    # Make the plot
    b1 = ax.bar(br1, list_tot_var1, color ="k", width = barWidth, edgecolor ='grey', label ='Total') 
    b2 = ax.bar(br2, list_forced_var1, color =CB_colors['blue'], width = barWidth, edgecolor ='grey', label ='Forced') 
    b3 = ax.bar(br3, list_intrinsic_var1, color ='darkgoldenrod', width = barWidth, edgecolor ='grey', label ='Intrinsic') 
    b4 = ax.bar(br4, list_res_var1, color ='gold', width = barWidth, edgecolor ='grey', label ='Bias') 
    
    # PLOT BAR OF --------------------------- EAI
    # set height of bar 
    list_tot_var2 = list_var_dep('EAI',COMP,'tot_var')
    list_forced_var2 = list_var_dep('EAI',COMP,'forced_var')
    list_intrinsic_var2 = list_var_dep('EAI',COMP,'intrinsic_var')
    list_res_var2 = list_var_dep('EAI',COMP,'res_var')
    
    # Set position of bar on X axis 
    br1 = np.arange(len(list_tot_var2)) + barWidth
    br2 = [x + 2*barWidth for x in br1] 
    br3 = [x + 2*barWidth for x in br2] 
    br4 = [x + 2*barWidth for x in br3] 
    
    # Make the plot
    b1_bis = ax.bar(br1, list_tot_var2,     color ="k", width = barWidth, alpha=0.7,edgecolor ='grey') 
    b2_bis = ax.bar(br2, list_forced_var2, color =CB_colors['blue'], width = barWidth,alpha=0.5,edgecolor ='grey') 
    b3_bis = ax.bar(br3, list_intrinsic_var2, color ='darkgoldenrod', width = barWidth, alpha=0.5,edgecolor ='grey') 
    b4_bis = ax.bar(br4, list_res_var2,      color ='gold', width = barWidth, alpha=0.5,edgecolor ='grey') 


    ax.set_xticks([r + barWidth for r in range(len(list_tot_var1))]) 
    ax.set_xticklabels(['0-50m', '0-100m','0-300m','0-700m','0-2000m','0-bottom'],fontsize=7) #, ) ,rotation=45, rotation_mode="anchor"


    
    return [(b1, b1_bis),(b2, b2_bis),(b3, b3_bis),(b4, b4_bis)]
```

```python
cm = 1/2.54  # centimeters in inches

fig, (ax0,ax1,ax2)= plt.subplots(1,3,figsize=(35*cm, 6*cm), facecolor='none', dpi=72*2) # 
# plt.subplots_adjust(wspace=0.1, hspace=0.7)

# bar chart
legend_handles = plot_bar_chart_dep("STERIC",ax0); ax0.set_yticks(np.arange(0,3.0,0.4)) 
plot_bar_chart_dep("THERMO",ax1); ax1.set_yticks(np.arange(0,3.0,0.4)) 
plot_bar_chart_dep("HALO",ax2);   ax2.set_yticks(np.arange(0,3.0,0.4)) #np.arange(0,0.5,0.1)) 

# ax0.legend()
l = ax0.legend(legend_handles, ['Total ES|ER', 'Forced ES|ER', 'Intrinsic ES|ER', 'Bias ES|ER'],\
               handler_map={tuple: HandlerTuple(ndivide=None)})

ax0.set_title("a) Steric")
ax1.set_title("b) Thermosteric")
ax2.set_title("c) Halosteric")

namo="synthetic_figure_atlantic_box_plot_depth_samey_axis.png"
ax0.set_ylabel(r'Variance [$cm^{2}$]')

# plt.show()
fig.savefig(diro+namo, facecolor=fig.get_facecolor(), bbox_inches='tight', transparent=True) #,edgecolor='none', bbox_inches='tight'
# plt.close(fig)
```

```python
# print les pourcentages
def array_var_dep(nexp,COMP,var_name):
    return_list=np.zeros(6)
    i=0
    for dep in ["50m", "100m", "300m", "700m", "2000m", "bottom"]: 
        var_ES_dep = xr.open_zarr(dir_base+nexp+"/"+COMP+"/var_terms_box_atl_0_"+dep+".zarr").variances
        return_list[i] = (var_ES_dep.sel(var_term=var_name).values)
        i+=1
        
    return return_list


# to get the numerical values
print("50m", "100m", "300m", "700m", "2000m", 'bottom')
for COMP in ['STERIC', 'THERMO','HALO']: #,'THERMO','HALO'
    print(COMP+ ' [%]' )
    list_tot_varES = array_var_dep('ES',COMP,'tot_var')
    list_tot_varEAI = array_var_dep('EAI',COMP,'tot_var')

    list_forced_varES = array_var_dep('ES',COMP,'forced_var')
    list_forced_varEAI = array_var_dep('EAI',COMP,'forced_var')

    list_intrinsic_varES = array_var_dep('ES',COMP,'intrinsic_var')
    list_intrinsic_varEAI = array_var_dep('EAI',COMP,'intrinsic_var')

    list_res_varES = array_var_dep('ES',COMP,'res_var')
    list_res_varEAI = array_var_dep('EAI',COMP,'res_var')

    list_internal_varEAI = list_intrinsic_varEAI - list_res_varEAI
    list_internal_varES = list_intrinsic_varES - list_res_varES


    # compute the percentage
    perc_tot_gain = ((list_tot_varEAI - list_tot_varES ) / list_tot_varES ) *100
    perc_forced_gain = ((list_forced_varEAI - list_forced_varES ) / list_forced_varES ) *100
    perc_intrinsic_gain = ((list_intrinsic_varEAI - list_intrinsic_varES ) / list_intrinsic_varES ) *100
    perc_res_gain = ((list_res_varEAI - list_res_varES ) / list_res_varES ) *100
    perc_internal_gain = ((list_internal_varEAI - list_internal_varES ) / list_internal_varES ) *100

    print("total: ",np.round(perc_tot_gain,2))
    print("forced: ",np.round(perc_forced_gain,2))
    print("intrinsic: ",np.round(perc_intrinsic_gain,2))
    print("bias: ",np.round(perc_res_gain,2))
    print("internal: ",np.round(perc_internal_gain,2))
```
