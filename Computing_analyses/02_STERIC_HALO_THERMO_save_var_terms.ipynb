# 1) save the four terms of the variance decomposition

```python
%%time
central_year = 1999
i=19 # compute just for the 39 year period

for nexp in ["EGAI", "EAI"]: #, "ES"
    print(nexp)
    for COMP in ["STERIC","HALO", "THERMO"]:
        print(COMP)
        varname = COMP+'_E'
        # dir_base = dir_sc + "/NEW_ENS_VAR_BUDGET/"+nexp+"/"+COMP+"/"
        dir_base = "/work/scratch-pw3/stajouri/STERIC_THERMO_HALO_2use/"+nexp+"/"+COMP+"/"
        
        for dep in ["50m", "100m", "300m", "700m", "2000m", "bottom"]: 
            print(dep)

            SSH = xr.open_zarr(dir_base+COMP+"_0_"+dep+"_2use_1980_2018.zarr")[varname]*100 # in cm
            
            # for i in np.arange(20):
            
            SSH_period = SSH.sel(year=slice(str(central_year-i),str(central_year+i)))
            TimeMean_SSH = SSH_period.mean('year').compute()
            EnsMean_SSH = SSH_period.mean('ens').compute()
            
            tot_var       = (((SSH_period - TimeMean_SSH)**2).mean("year") ).mean("ens")
            forced_var    = ((EnsMean_SSH - EnsMean_SSH.mean("year"))**2).mean("year")
            intrinsic_var = (((SSH_period - EnsMean_SSH)**2).mean("ens")).mean('year')
            res_var       = ((TimeMean_SSH - TimeMean_SSH.mean('ens'))**2).mean('ens')
            
            Variance_terms = xr.Dataset(
            data_vars=dict(
                tot_var=(["y", "x"], tot_var.data),
                forced_var=(["y", "x"], forced_var.data),
                intrinsic_var=(["y", "x"], intrinsic_var.data),
                res_var=(["y", "x"], res_var.data) ))
            Variance_terms.to_zarr(dir_base+"var_terms_"+str(1+ i*2)+"_0_"+dep+".zarr", mode='w', zarr_version=2)
```

# 2) Save the GM var for each term for each time period length

```python
diri="/gpfswork/rech/cli/rcli002/eORCA025.L75/eORCA025.L75-I/"
mesh_hgr = xr.open_dataset(diri+'mesh_hgr.nc').squeeze()
tmask = mesh_hgr.tmask[0].drop_vars('nav_lev').drop_vars('time_counter')

e1t = mesh_hgr.e1t.fillna(0)
e2t = mesh_hgr.e2t.fillna(0)

# constructing the weight
bt = (e1t * e2t).load()# area of each cell
# getting the weights to be applied to every grid cell
# masking the land cells
oceanArea = (bt*tmask)  #tmask : 1 on ocean, 0 on land
totalOceanArea = oceanArea.sum()
Weight = (oceanArea/totalOceanArea).compute().drop_vars('time_counter')
Weight.sum()
```

```python
%%time
dep ="bottom"

for nexp in ["ES"]:
    for COMP in ["STERIC","HALO", "THERMO"]:
        print(COMP)
        my_array = np.zeros((4,20))
        
        dir_base = dir_sc + "/NEW_ENS_VAR_BUDGET/"+nexp+"/"+COMP+"/"
    
        for i in np.arange(20):
            # print(str(1+ i*2)) 
            var_ds_corr = xr.open_zarr(dir_base+"var_terms_"+str(1+ i*2)+"_0_"+dep+".zarr")
            
            tot_var_corr = var_ds_corr.tot_var
            forced_var_corr = var_ds_corr.forced_var
            intrinsic_var_corr = var_ds_corr.intrinsic_var
            res_var_corr = var_ds_corr.res_var
        
            tot_var_corrGM       = (tot_var_corr * Weight).sum(('x','y'))
            forced_var_corrGM    = (forced_var_corr * Weight).sum(('x','y'))
            intrinsic_var_corrGM = (intrinsic_var_corr * Weight).sum(('x','y'))
            res_var_corrGM       = (res_var_corr * Weight).sum(('x','y'))
        
            my_array[0,i] = tot_var_corrGM
            my_array[1,i] = forced_var_corrGM
            my_array[2,i] = intrinsic_var_corrGM
            my_array[3,i] = res_var_corrGM
        
        Variance_terms = xr.Dataset(
            data_vars=dict(
                global_var=(["var_term", "period_length"], my_array)),
            coords=dict(
                var_term=["total", "forced", "intrinsic","bias"],
                period_length=np.arange(1,41,2)))
        
        Variance_terms.to_zarr(dir_base+"global_var_terms.zarr", mode='w')
```
