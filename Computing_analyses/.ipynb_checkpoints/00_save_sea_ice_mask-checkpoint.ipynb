# save sea ice mask

```python
dir_work = os.environ["WORK"]
```

```python
%%time
prefix = "eORCA025.L75-IMHOTEP"
chunk_size = {"x":133,"y":320}

for nexp in ["ES", "EAI", "EGAI"]: 
    dir_base = dir_work+"/ETUDE3/ENS_VAR_BUDGET/raw_data/"+nexp+"/"
    # sea ice mask of ice concentration
    
    ii=1
    totb=[]
    
    for Emb in ['001','002','003','004','005','006','007','008','009','010']:
        diridatref="/gpfsstore/rech/cli/rcli002/eORCA025.L75/"+prefix+"."+nexp+"."+Emb+"-S/1y/"
        simsk15 = xr.open_mfdataset(diridatref+'????/*_icemod.nc', chunks = chunk_size, parallel=True).simsk15.squeeze()
        
        totb.append(simsk15.fillna(0).assign_coords({'ens':ii}))
        ii+=1
        
    SIMASK15_E= xr.concat(totb, 'ens').sel(time_counter=slice("1980","2018"))
    
    # create the mask of 0 and 1 as soon as sea ice appears.
    SIMASK15_E2apply = SIMASK15_E.where(SIMASK15_E != 0, 2)
    SImask_finial = SIMASK15_E2apply.where(SIMASK15_E2apply==2,0) /2
    
    SImask_finial.to_dataset(name="SIMASK15_E").to_zarr(dir_base+nexp+"_SIMASK15_E1980_2018.zarr", mode='w')
```
