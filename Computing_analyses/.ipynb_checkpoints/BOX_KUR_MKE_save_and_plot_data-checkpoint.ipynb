# Compute MKE for each ensemble member ii and then do the ensemble mean and save

```python
dir_ww_out = dir_work + "/ETUDE3/ENS_VAR_BUDGET/raw_data/"
chunk_size = {"x":133,"y":320}
dir_out_sc = dir_sc + "/ENS_VAR_BUDGET/"
fo="1m"
for nexp in ["ES","EAI", "EGAI"]: #"EAI","EGAI"
    ii=1
    ens_U=[]
    ens_V=[]

    for Emb in ['001','002','003','004','005','006','007','008','009','010']:
        dir_ww_data = dir_ww_out+nexp+"/MKE/"
        
        U_bar = xr.open_zarr(dir_ww_data+Emb+"_Usurf_"+fo+"_bar.zarr").vozocrtx
        V_bar = xr.open_zarr(dir_ww_data+Emb+"_Vsurf_"+fo+"_bar.zarr").vomecrty
        
        ens_U.append(U_bar.assign_coords({'ens':ii}))
        ens_V.append(V_bar.assign_coords({'ens':ii}))

        ii+=1
        
        U_E= xr.concat(ens_U, 'ens')
        V_E= xr.concat(ens_V, 'ens')
    
    # compute MKE
    MKE = 0.5 * (U_E**2 + V_E**2)

    # compute the ensemble mean
    MKE_Emean = MKE.mean('ens')

    # save
    MKE_Emean.to_dataset(name="MKE_Emean").to_zarr(dir_out_sc+nexp+"/MKE/"+"MKE_Emean_"+fo+".zarr", mode='w')
    
```
